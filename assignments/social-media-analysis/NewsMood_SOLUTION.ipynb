```python
# Dependencies
import tweepy
import numpy as np
import pandas as pd
from datetime import datetime
import matplotlib.pyplot as plt
from matplotlib import style
style.use('ggplot')

# Import and Initialize Sentiment Analyzer
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
analyzer = SentimentIntensityAnalyzer()

# Twitter API Keys
from tweepy_config import (consumer_key, 
                    consumer_secret, 
                    access_token, 
                    access_token_secret)

# Setup Tweepy API Authentication
auth = tweepy.OAuthHandler(consumer_key, consumer_secret)
auth.set_access_token(access_token, access_token_secret)
api = tweepy.API(auth, parser=tweepy.parsers.JSONParser())
```

# BBC results

```python
# BBC user handle = @BBC

# Target Account
target_user = "@BBC"

# Counter
counter = 1

# Variables for holding sentiments
bbc_sentiments = []

# Variable for max_id
oldest_tweet = None

# Loop through 5 pages of tweets (total 100 tweets)
for x in range(5):

    # Get all tweets from home feed
    public_tweets = api.user_timeline(target_user, max_id = oldest_tweet)

    # Loop through all tweets 
    for tweet in public_tweets:
        
        # Run Vader Analysis on each tweet
        results = analyzer.polarity_scores(tweet["text"])
        compound = results["compound"]
        pos = results["pos"]
        neu = results["neu"]
        neg = results["neg"]
        
        # Get Tweet ID, subtract 1, and assign to oldest_tweet
        oldest_tweet = tweet['id'] - 1
        
        # Add sentiments for each tweet into a list
        bbc_sentiments.append({"Date": tweet["created_at"], 
                           "Compound": compound,
                           "Positive": pos,
                           "Negative": neu,
                           "Neutral": neg,
                           "Tweets Ago": counter})
        
        # Add to counter 
        counter += 1
```

```python
# Convert sentiments to DataFrame
bbc_df = pd.DataFrame.from_dict(bbc_sentiments)
bbc_df.head()
```

# The first plot

* Be a scatter plot of sentiments of the last **100** tweets sent out by each news organization, ranging from -1.0 to 1.0, where a score of 0 expresses a neutral sentiment, -1 the most negative sentiment possible, and +1 the most positive sentiment possible.
* Each plot point will reflect the _compound_ sentiment of a tweet.
* Sort each plot point by its relative timestamp.

```python
# Create plot

bbc, = plt.plot(bbc_df['Tweets Ago'], bbc_df['Compound'], marker="o", linewidth=0 ,alpha=0.8, c='k', label="BBC")
cbs, = plt.plot(cbs_df['Tweets Ago'], cbs_df['Compound'], marker="^", linewidth=0 ,alpha=0.8, c='g', label="CBS")
cnn, = plt.plot(cnn_df['Tweets Ago'], cnn_df['Compound'], marker="s", linewidth=0 ,alpha=0.8, c='b', label="CNN")
fox, = plt.plot(fox_df['Tweets Ago'], fox_df['Compound'], marker="x", linewidth=0 ,alpha=0.8, c='r', label="Fox")
nyt, = plt.plot(nyt_df['Tweets Ago'], nyt_df['Compound'], marker="d", linewidth=0 ,alpha=0.8, c='y', label="NYT")

# Incorporate the time
now = datetime.now()
now = now.strftime("%Y-%m-%d %H:%M")

# Plot properties
plt.rcParams["figure.figsize"] = (20,10)
plt.title(f"Sentiment Analysis of Tweets ({now})")
plt.xlim(0,101)
plt.ylim(-1,1)
plt.ylabel("Tweet Polarity")
plt.xlabel("Tweets Ago")
plt.legend(handles=[bbc,cbs,cnn,fox,nyt], loc="best")
plt.savefig("sentiment_analysis_time.png")
plt.show()
```

# The second plot 

will be a bar plot visualizing the _overall_ sentiments of the last 100 tweets from each organization. For this plot, you will again aggregate the compound sentiments analyzed by VADER.

```python
# Calculate averages
bbc_avg = round(np.average(bbc_df['Compound']),3)
cbs_avg = round(np.average(cbs_df['Compound']),3)
cnn_avg = round(np.average(cnn_df['Compound']),3)
fox_avg = round(np.average(fox_df['Compound']),3)
nyt_avg = round(np.average(nyt_df['Compound']),3)

# Create the list of averages
avg = [['BBC',bbc_avg],['CBS',cbs_avg],['CNN',cnn_avg],['Fox',fox_avg],['NYT',nyt_avg]]

# Convert to a dataframe
avg_df = pd.DataFrame(avg, columns=['News', 'Tweet Polarity'])
avg_df
```

```python
# Create bar chart
x_axis = np.arange(len(avg_df))
plt.bar(x_axis, avg_df['Tweet Polarity'], color=("k","g","b","r","y"), align="center")

# Create the ticks for bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, avg_df['News'])

ymin = min(avg_df['Tweet Polarity'])-.025
ymax = max(avg_df['Tweet Polarity'])+.025

# Plot properties
plt.rcParams["figure.figsize"] = (20,10)
plt.title(f"Aggregate Sentiment Analysis of Tweets ({now})")
plt.xlim(-0.75,max(x_axis)+0.75)
plt.ylim(ymin,ymax)
plt.ylabel("Aggregate Polarity")
plt.xlabel("News")
plt.savefig("sentiment_analysis_aggregate_skewed.png")
plt.show()
```

```python
# Create bar chart (not skewed)
x_axis = np.arange(len(avg_df))
plt.bar(x_axis, avg_df['Tweet Polarity'], color=("k","g","b","r","y"), align="center")

# Create the ticks for bar chart's x axis
tick_locations = [value for value in x_axis]
plt.xticks(tick_locations, avg_df['News'])

# Plot properties
plt.rcParams["figure.figsize"] = (20,10)
plt.title(f"Aggregate Sentiment Analysis of Tweets ({now})")
plt.xlim(-0.75,max(x_axis)+0.75)
plt.ylim(-1,1)
plt.ylabel("Aggregate Polarity")
plt.xlabel("News")
plt.savefig("sentiment_analysis_aggregate.png")
plt.show()
```
