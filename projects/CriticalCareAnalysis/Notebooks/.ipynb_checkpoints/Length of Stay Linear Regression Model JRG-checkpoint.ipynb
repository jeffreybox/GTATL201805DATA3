```python
import pandas as pd
import numpy as np
```

```python
labs = pd.read_csv('../Resources/labs.csv')
```

```python
admissions_survived = pd.read_csv('../Resources/admissions_survived.csv')
```

```python
labs.head()
```

```python
admissions_survived.head()
```

```python
survived_df = pd.merge(admissions_survived,labs, on='HADM_ID')
```

```python
survived_df.head()
```

```python
survived_df.columns
```

```python
y = survived_df['LENGTH_OF_STAY']
```

```python
survived_df = survived_df.drop(['ROW_ID', 'SUBJECT_ID_x', 'HADM_ID', 'ADMITTIME', 'DISCHTIME', 'DISCHARGE_LOCATION',
                         'RELIGION', 'ETHNICITY', 'EDREGTIME', 'EDOUTTIME', 'HOSPITAL_EXPIRE_FLAG',
       'HAS_CHARTEVENTS_DATA', 'LENGTH_OF_STAY', 'SUBJECT_ID_y', 'LANGUAGE'], axis = 1)
```

```python
X = pd.get_dummies(survived_df)
```

```python
X.head()
```

```python
y = y.values.reshape(-1, 1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

y_scaler = StandardScaler().fit(y_train)
```

```python
y_train_scaled = y_scaler.transform(y_train)
y_test_scaled = y_scaler.transform(y_test)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train_scaled)
```

```python
from sklearn.metrics import mean_squared_error

predictions = model.predict(X_test)
MSE = mean_squared_error(y_test_scaled, predictions)
r2 = model.score(X_test, y_test_scaled)

print(f"MSE: {MSE}, R2: {r2}")
```

```python
from sklearn.feature_selection import f_regression

scores, pvalues = f_regression(X_train, y_train_scaled)
```

```python
for i in range(len(pvalues)):
    print(f'{i}.{X.columns[i]}: {pvalues[i]}')
```

```python
drop_features = []
for i in range(len(pvalues)):
    if not pvalues[i] < 0.05:
        drop_features.append(X.columns[i])
```
