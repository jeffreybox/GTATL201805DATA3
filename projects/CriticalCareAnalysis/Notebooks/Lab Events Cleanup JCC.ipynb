```python
import pandas as pd
import numpy as np
```

```python
blood_gas = pd.read_csv("../../NewResources/blood_gas.csv")
chemistry = pd.read_csv("../../NewResources/chemistry.csv")
hematology = pd.read_csv("../../NewResources/hematology.csv")

blood_gas = blood_gas[['SUBJECT_ID','HADM_ID','VALUENUM','LABEL','CATEGORY']]

chemistry = chemistry[['SUBJECT_ID','HADM_ID','VALUENUM','LABEL','CATEGORY']]

hematology = hematology[['SUBJECT_ID','HADM_ID','VALUENUM','LABEL','CATEGORY']]

blood_gas.head()
```

```python
blood_gas.loc[blood_gas['LABEL']=='Hemoglobin','LABEL'] = 'Hemoglobin_Blood_Gas'
blood_gas.loc[blood_gas['LABEL']=='Glucose','LABEL'] = 'Glucose_Blood_Gas'
chemistry.loc[chemistry['LABEL']=='Glucose','LABEL'] = 'Glucose_Chemistry'
hematology.loc[hematology['LABEL']=='Hemoglobin','LABEL'] = 'Hemoglobin_Hematology'
```
